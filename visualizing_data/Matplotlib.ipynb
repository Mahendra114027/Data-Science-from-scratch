# Matplotlib

### Line charts

```python
from matplotlib import pyplot as plt
import random
```

```python
years = [1950+(i*10) for i in range(10)]
gdp = [300, 500, 1000, 2000, 3600, 5900, 7000, 10200, 15000, 20100]
print(years, gdp)
```

```python
plt.title('Nominal GDP')
plt.ylabel("Billions of $")
plt.xlabel("Years")
plt.plot(years, gdp, color='green', marker='o', linestyle='solid')
```

### Bar charts

Good choice when used to show how a certain quantity varies within discrete set of items

```python
movies = ["Annie Hall", "Ben-hur", "Casablanca", "Gandhi", "West side story"]
num_oscars = [5, 11, 3, 8, 10]
```

```python
plt.ylabel("# of Academy awards")
plt.title("Movies")
plt.xticks(range(len(movies)), movies)
plt.bar(range(len(movies)), num_oscars)
```

Good choice for plotting histograms

```python
from collections import Counter
grades = [random.randrange(0, 100) for _ in range(20)]
decile = lambda grade: grade
histogram = Counter(decile(grade) for grade in grades)
```

```python
# FIXME: Bar width is not proper
plt.axis([-5, 105, 0, 5]) # x-axis = -5 to 105 and y-axis = 0 to 5
plt.xticks([10*i for i in range(10)])
plt.xlabel("Decile")
plt.ylabel("# of students")
plt.title("Distribution of Exam 1 grades")
plt.bar(histogram.keys(), histogram.values())
```

### Line charts with legends

```python
variance = [2**i for i in range(9)]
bias_squared = variance[::-1]
total_error = [x+y for x, y in zip(variance, bias_squared)]
xs = [i for i, _ in enumerate(variance)]
```

```python
plt.xlabel("Model complexity")
plt.title("The bias-variance tradeoff")
plt.plot(xs, variance, 'g-', label='var')
plt.plot(xs, bias_squared, 'r-.', label='bias^2')
plt.plot(xs, total_error, 'b:', label='error')
plt.legend(loc=9)
```

### Scatter plots

Useful when representing the relationship between two pairs of data

```python
friends = [70, 65, 72, 63, 71, 64, 60, 64, 67]
minutes = [175, 170, 205, 120, 220, 130, 105, 145, 190]
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i']

plt.scatter(friends, minutes)

for label, friend_count, minute_count in zip(labels, friends, minutes):
    plt.annotate(label,
                 xy=(friend_count, minute_count),
                 xytext=(5, -5),
                 textcoords='offset points')
plt.xlabel("# of friends")
plt.ylabel("daily minutes spent on the sites")
```
